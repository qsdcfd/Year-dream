**문제 해석**

- 재귀 함수

   - 0을 발견하면 재귀함수로 그래프에서 구멍 채우기(1)

   - 새로운 구멍 찾아서 개수를 카운트

<br>

- 큐 자료구조

   - 굳이 방문확인 할 필요없는 BFS
   
- 0을 1로 바꿀 것이기에 0과 1을 True,False역할

```python
import sys
from collections import deque

input  = sys.stdin.readline #입력값 한 줄 씩 읽기

N,M = map(int, input().split()) # 가로, 세로
direc = [[1,0], [-1,0], [0,1], [0,-1]]# 상하좌우방향 할당

graph = [list(map(int, list(input().strip()))) for _ in range(N)] #그래프탐색

def BFS1(row, col): ##BFS row * col
    queue =  deque()
    queue.append([row, col])
    graph[row][col] = 1 #  1인 거 찾기

    while queue:
        now_r, now_c = queue.popleft() #row, col을 보면서 0,1찾기 그리고 popleft
        for r, c in direc:
            new_r = now_r + r #row쪽 이동
            nrw_c = now_c + c #col쪽 이동
            if 0 <= new_r < N and 0 <=new_c < M:
                if graph[new_r][new_c] ==0: #방문처리 배열을 만들어진 않는다.
                    queue.append([new_r, new_c])
                    graph[new_r][new_c] = 1 #큐에 중복된 좌표를 넣지 않기 위해서
    return True

count = 0

for i in range(N):
    for j in range(M):
        if graph[i][j] ==0 : #새로운 구멍 발견 시 
            BFS1(i,j) #구명 메꾸기
            count +=1

print(count)


```
