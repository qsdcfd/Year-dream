### ROC 와 AUC

- 학습기는 테스트 세트를 위해서 실수값 혹은 확률 예측값 계산

- 해당 예측값과 하나의 분류 임계치와 비교

- 임계치보다 크면 양성값, 작으면 음성값

- 위와 같은 방식은 해당 순서에 하나의 차단점을 기준으로 두어서 샘플을 양분하는 과정과 유사

   - 배열 앞쪽에 위치한 부분을 차단점으로 설정할 경우, 정밀도가 중요

   - 배열 뒤쪽에 위치한 부분을 차단점으로 설정할 경우, 재현율이 더 중요

   - **배열에 순서 자체의 품질에 따라ㅓ 각 학습기의 일반화 성능 결정**

<br>

**ROC곡선**

image.png

- 수신기 조작 특성의 약자이다.

- 적군의 레이더 신호를 분석하는 기술로 활용되었다.

- y축에는 True Positive Rate

- x축에는 False Positive Rate
image.png

-  데이터의 수가 많지 않을 것이기에 b와 같은 근사값을 기반으로 그려질 것이다.

- 그래프 만드는 과정

   - m+개 양성값과 m-개 음성값이 있다고 가정
   - 학습기 예측 결과에 기반을 둔 샘플을 오름차순으로 정렬
   - 참 양성률과 거짓 양성률은 모두 0
   - 좌표(0,0)에 점을 찍고, 임계치를 각 샘플의 예측값으로 정함
   - 각 샘플을 양성값으로 분류
   - 값 좌표를 (x,y)라고 설정
   - 참 양성률은(x,y+1/m+)
   - 거짓 양성률은 (x+1/m-, y)
   - 직선을 이용하여 각 점 잇기

- 학습기를 비교할 때 포함을 하는 곡선이 우수한 학습기

<br>

### AUC

- 두 학습기 ROC 곡선에 교차가 발생하면 우열을 가르기 힘들다.

- 그래서, ROC곡선 아래의 면적 비교

- AUC는 ROC곡선 아래 각 부분의 면적을 모두 더해서 구한다.

image.png

- AUC는 샘플 예측의 배열 순서 품질을 고려하기에 순서 오차와 긴밀한 관계가 있다.

   - 양성값의 예측이 음성값 보다 작게 나오면 벌점 부여
   - 같다면 0.5개의 벌점을 매긴다.
   - 즉, 거짓 양성률을 나타낸다.
      - AUC = 1-lrank
