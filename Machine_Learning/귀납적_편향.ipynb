### 귀납적 편향

**Inrto**

- 학습된 모델은 하나의 가설에 대응을 한다.

- 대응된 모델이 새로운 샘플을 만나게 되면 서로 다른 출력이 나오게 된다.(if 훈련 데이터 세트와 일치한 가설이 있을 경우)

- 이러한 상황 속에서 우리의 고민은?

   - 어떤 모델과 어떤 가설을 이용해야하나?

<br>

**등장배경**



- 편향이 없는 훈련 샘플 데이터

- 여러  가설 중에서 무슨 가설이 좋은 것이 판단 어려움

- 모델도 혼란이 가중되어서 확실한 학습결과 생성 어려움

- 그러나, 구체적인 학습 알고리즘은 하나의 모델을 생성해야한다.


image.png

**귀납적 편향의 필요성**

- 학습 알고리즘이 방대한 가설 공간에서 활용 가능성이 높은 가설이 선택되도록 도와줌

    - 휴리스틱 방법

    - 가치관 


- 학습 알고리즘이 내놓은 모델 중 어떤 모델이 더 좋은가 관한 가설 대응

- 실제 성립 여부는 문제와 얼마나 잘 맞는지는 결국 알고리즘의 좋은 성능 가지는 여부에 의해서 결정됨


<br>

**오컴의 면도날**

- 편향에 대한 옳고 그름을 판단하는 원칙

- 다수의 가설이 관측된 결과가 일치하다면 가장 간단한 것을 선택

- 다른 해석도 나올 가능성이 있기에 오컴의 면도날이 완벽한 원칙은 아니다.

### 공짜 점심은 없다(NFL)정리

- 균등분포일 때 해당하는 내용임을 미리 알린다.

- 무슨 알고리즘이 좋은지는 찍기에 가깝다.

- 총 오차는 알고리즘과 무관

   - 여러 학습 알고리즘이 똑똑하던 멍청하던 그들의 기대성능은 같다.


- 교훈

   - 모든 잠재적 문제를 고려하면 모든 학습 알고리즘이 동등하게 좋을 것이다.

   - 만약 알고리즘의 좋고 나쁨을 비교하면 특정하고 구체적인 학습 문제와 비교해야한다.

   - 문제에 따라 성능이 달라지는 알고리즘도 있다.

   - 그러한 이유로 학습 알고리즘의 본연의 귀납적 편향과 문제가 맞지 않을 수 있다.
