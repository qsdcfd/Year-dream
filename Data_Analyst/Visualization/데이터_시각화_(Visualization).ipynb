# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 10)
```

```python
y = x ** 2
```

```python
plt.plot(x, y, 'g')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.7, 1])

axes.plot(x, y, 'r')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
plt.plot(x, y, 'g')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
plt.plot(y, x, 'b')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) # 좌우 위치, 위아래 위치, 너비, 높이 순
axes2 = fig.add_axes([0.2,0.6,0.3,0.3])

axes1.plot(x, y, 'r')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title 1')

axes2.plot(y, x, 'b')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title 2')
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'g')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y, 'r')
axes[1].plot(y,x, 'b')

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Title_1')

axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Title_2')
```

```python
fig.savefig('my_name.png')
```

# Seaborn - Distribution Plot

```python
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

### Distribution Plot

```python
sns.distplot(data['total_bill'])
```

```python
sns.distplot(data['total_bill'], kde = False)
```

```python
sns.distplot(data['total_bill'], kde = False, bins = 10)
```

### Joint Plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'reg')
```

### Pair Plot

```python
sns.pairplot(data, hue = 'smoker')
```

### Rug Plot

```python
sns.rugplot(data['total_bill'])
```

# Seaborn - Categorical Plot

### Bar Plot

```python
sns.barplot(x = 'smoker', y= 'total_bill', data = data)
```

```python
sns.barplot(x = 'smoker', y= 'total_bill', data = data, estimator = np.std)
```

### Count Plot

```python
sns.countplot(x = 'sex', data = data)
```

```python
sns.countplot(data['sex'])
```

### Box Plot

```python
sns.boxplot(x= 'day', y = 'total_bill', data = data, palette = 'Set2')
```

```python
sns.boxplot(x= 'day', y = 'total_bill', data = data, palette = 'Set2', hue = 'smoker')
```

### Violinplot

```python
sns.violinplot(x= 'day', y = 'total_bill', data = data, palette = 'Set2', hue = 'smoker', split = True)
```
